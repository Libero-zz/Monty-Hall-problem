```python
import random

# 我這裡偷懶了，早上的程式DOORS可以亂改，但這裡的程式只有DOORS = 3的時候才會對喔
DOORS = 3
PLAY_TIMES = 100000

win_times_player_no_change = [0, 0, 0]
win_times_player_change = [0, 0, 0]
win_times_player_not_in_car = [0, 0, 0]

# 主持人看一看答案 直接把C開了 結果丙共辜
# 這句話的解釋法，是整題關鍵
for i in range(PLAY_TIMES):
    # 建構ABC三扇門
    door = list(range(DOORS))
    
    # 甲、乙、丙三個人選擇答案(不能重複)
    player = list(range(DOORS))
    pick = list(range(DOORS))
    random.shuffle(pick)

    ############# 狀況一，主持人隨機去掉沒中的一扇門 #############
    # 主持人手中握有答案
    answer = random.choice(door)

    # 主持人隨機去掉沒中的一扇門
    remove_choices = list(range(DOORS))
    remove_choices.remove(answer)
    remove_choice = random.choice(remove_choices)
    door.remove(remove_choice)

    # 因此某位玩家也出局囉
    remove_index = pick.index(remove_choice)
    player.pop(remove_index)
    pick.pop(remove_index)

    ##########################

    # 然後問甲乙要不要換門
    for i in range(DOORS):
        # 以下切換到玩家視角
        # 倖存的玩家不換門
        if i in player:
            player_index = player.index(i)
            # a.我不換門
            if pick[player_index] == answer:
                win_times_player_no_change[i] += 1
            
            # b.我換門，即如果我原本猜錯的話，我之後會答對
            if pick[player_index] != answer:
                win_times_player_change[i] += 1
        else:
            # 我沒上車啊
            win_times_player_not_in_car[i] += 1
    
print("甲不換贏的機率:", win_times_player_no_change[0]/PLAY_TIMES*100)
print("甲換贏的機率：",win_times_player_change[0]/PLAY_TIMES*100)
print("甲第一輪就被刷掉的機率",win_times_player_not_in_car[0]/PLAY_TIMES*100)
```
